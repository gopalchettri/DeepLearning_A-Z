# Artificial Neural Network - Customer Churn

```python
# importing required libraries

import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import StandardScaler
```

```python
# importing the dataset
location = "C:/DevelopmentPlayground/Datasets/Churn_Modelling.csv"
dataset = pd.read_csv(location)
```

```python
dataset.head()
```

## EDA

```python
dataset.describe().T
```

```python
# checking if there are any null values

dataset.isnull().values.any()
```

```python
dataset.shape
```

```python
# Splitting the dataset into features (X) and label (y)

X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
X.shape
```

```python
# Encoding the Gender column

labelencoder_gender = LabelEncoder()
X[:, 2] = labelencoder_gender.fit_transform(X[:, 2])
```

```python
print(X)
print(X.shape)
```

```python
# One Hot encoding the Geography Column

columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(columnTransformer.fit_transform(X))
```

```python
print(X)
print(X.shape)
```

```python
# Splitting the dataset into training and testing set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Building the ANN

```python
# importing the keras libraries

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# initialing the ANN

classifier = Sequential(name="ANN-classifier")

# adding the first hidden layer and input dimension. input_dim = 12 becauze we have 12 columns

classifier.add(Dense(input_dim=12, units=6, activation='relu', kernel_initializer='uniform', name="Hidden-Layer-1"))

# adding the second hidden layer

classifier.add(Dense(units=6, activation='relu', kernel_initializer='uniform', name="Hidden-Layer-2"))

# adding the output layer

classifier.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform', name="Output-Layer"))
```

```python
# compiling the ANN 

classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# fitting the ANN to the Training set

classifier.fit(X_train,y_train, batch_size=32, epochs=100)
```

```python
classifier.summary()
```

## Making the predictions and evaluating the model


```python
# predicting the Test set results

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
score = accuracy_score(y_test, y_pred)
```

```python
print('confusion matrix :\n', cm)
print('\naccuracy score :\n', score)
```

```python
# Evaluate the model
scores = classifier.evaluate(X_train, y_train)
print('model_metrices and scores :', classifier.metrics_names[1], scores[1]*100)
```

## Evaluating and  Improving and Tuning the ANN

##  Evaluating the ANN

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(input_dim=12, units=6, kernel_initializer='uniform', activation='relu', name='hidden-layer-1'))
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', name='hidden-layer-2'))
    classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid', name='output-layer'))
    classifier.compile(optimizer=optimizer , loss='binary_crossentropy', metrics=['accuracy'])
    return classifier
```

```python
import time
start_time = time.time()

# creating model using KerasClassifier
classifier = KerasClassifier(build_fn=build_classifier(), batch_size=10, epochs=100, verbose=1)
accuracies = cross_val_score(estimator=classifier, X = X_train, y= y_train, cv=10)

print('Time taken :', time.time() - start_time)
```

```python
mean = accuracies.mean()
variance = accuracies.std()
```

```python
print('Mean :', mean)
print('Variance :', variance)
```

## Tuning the ANN 

```python
from sklearn.model_selection import GridSearchCV

classifier = KerasClassifier(build_fn= build_classifier)

# batch_size = [10, 20]
# epochs = [10, 50]
# optimizer = ['adam','rmsprop']
# param_grid = dict(optimizer=optimizer,batch_size=batch_size, epochs=epochs)

parameters = {'batch_size': [25, 32],
              'epochs': [100, 200],
              'optimizer': ['adam', 'rmsprop']}
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 5)

# grid_search = GridSearchCV(estimator=classifier, param_grid=param_grid, cv=3)
```

```python

start_time = time.time()

grid_search = grid_search.fit(X_train, y_train)

print('Total time took :', time.time() - start_time)
```

```python
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
