## Multiclass Classification with the Iris Flowers Dataset -  ANN classification

```python
# importing required packages

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras import utils

from sklearn.model_selection import StratifiedKFold # for k-fold validation
from sklearn.model_selection import KFold

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import LabelEncoder

from sklearn.pipeline import Pipeline
```

```python
# fix random seed for reproducibility

seed = 7
np.random.seed(seed)
```

```python
# importing the dataset

location = "C:/DevelopmentPlayground/Datasets/iris.csv"

dataset = pd.read_csv(location)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
# Spliting the dataset into X and Y

X = dataset.iloc[:, 1:5].values
Y = dataset.iloc[:, -1].values
```

```python
# print first 5 X and Y

print(X[0:5])
print(Y[0:5])
```

```python
print(X.shape)
print(Y.shape)
```

```python
# Encoding the class values as integers

lblencoder = LabelEncoder()
lblencoder.fit(Y)
encoded_Y = lblencoder.transform(Y)

print('Encoded Y :', encoded_Y)
```

```python
# Convert the integers to dummy variables i.e., one hot encoding

Y = utils.to_categorical(encoded_Y)
# Y = utils.to_categorical(encoded_Y)
```

```python
print('Y : \n', Y[0:5])
```

```python
print('type(X) :', type(X))
print('type(Y) :',type(Y))
```

```python
print('X Shape :', X.shape)
print('Y Shape :', Y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=seed)
```

```python
# creating the model
def create_baseline_model(optimizer='adam', kernel='glorot_uniform'):
    model = Sequential()
    model.add(Dense(units=8, input_dim=4, activation='relu', kernel_initializer = kernel, name='Hidden-Layer-1'))
    model.add(Dense(units=3, activation='softmax', kernel_initializer = kernel, name='Output-Layer'))
    # compile the model
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn= create_baseline_model, verbose=1)
```

##  Grid SearchCV

```python
# grid search epoch, batch size and optimizer

optimizers = ['adam', 'rmsprop']
kernels = ['glorot_uniform', 'normal', 'uniform']
epochs = np.array([50, 100, 150])
batches = np.array([5, 10, 20])
```

```python
param_grid = dict(optimizer = optimizers, epochs=epochs, batch_size=batches, kernel=kernels)
```

```python
grid=GridSearchCV(estimator= model, param_grid=param_grid)
```

```python
grid_result = grid.fit(X_train, y_train)
```

```python
print('Best Score :', grid_result.best_score_)
print('Best Parameters :', grid_result.best_params_)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print("\n\n")

means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
