```python
# importing required libraries

import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from keras.callbacks import ModelCheckpoint

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

from sklearn.model_selection import GridSearchCV

from sklearn.pipeline import Pipeline
import h5py
import time

```

```python
# function to convert seconds to hour min and sec

def convert(seconds):
    min, sec = divmod(seconds, 60)
    hour, min = divmod(min, 60)
    return "%d:%02d:%02d" % (hour, min, sec)
```

```python
# fix the random seed for reproducibility

seed =7
np.random.seed(seed)
```

```python
# loading dataset 
location = "C:/DevelopmentPlayground/Datasets/sonar_csv.csv"

dataset = pd.read_csv(location)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
# Spliting the dataset into X and Y

X = dataset.iloc[:, 0:60].values
Y = dataset.iloc[:, -1].values
```

```python
# print first 2 X and Y

print(X[0:2])
print(Y[0:2])
```

```python
# Encoding the class values as integers

lbl_encoder = LabelEncoder()
lbl_encoder.fit(Y)
ecoded_Y = lbl_encoder.transform(Y)
```

```python
# print encoded Y

print(ecoded_Y)
```

```python
# splitting the data into training and testing set

X_train, X_test, y_train, y_test = train_test_split(X, ecoded_Y, test_size=0.33, random_state=seed)
```

```python
# creating base line model

def create_baseline_model(optimizer='adam', kernel='glorot_uniform'):
    model = Sequential()
    model.add(Dense(units=60, input_dim=60, kernel_initializer=kernel, activation='relu', name='Hidden-Layer-1'))
    model.add(Dense(units=1, kernel_initializer=kernel, activation='sigmoid', name='Output-Layer-1'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

## Pipeline

```python
kears_estimator = KerasClassifier(build_fn=create_baseline_model, validation_split=0.33, verbose=1)

pipe = Pipeline([('ss', StandardScaler()), 
                ("kc", kears_estimator)])

sorted(pipe.get_params().keys())
```

```python
pipe.fit(X_train, y_train)
```

```python
print('Training set score: ' + str(pipe.score(X_train,y_train)))
print('Test set score: ' + str(pipe.score(X_test,y_test)))
```

## cross validation using cross_val_score

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```

## Grid SearchCV

```python
# grid search epoch, batch size and optimizer

parameters = {
    'kc__optimizer': ('adam', 'rmsprop'),
    'kc__kernel': ('glorot_uniform', 'normal', 'uniform'), # kernel
    'kc__epochs': [10,20,30,50],
    'kc__batch_size': [5, 10, 20],
}
```

```python
kfold =  StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
grid_search=GridSearchCV(pipe, parameters, cv= kfold, verbose=1)
```

```python
import time
start_time = time.time()

grid_result = grid_search.fit(X_train, y_train)

training_time = time.time() - start_time
print('\n Total Training Time taken :', convert(training_time))
```

## Access the best set of parameters

```python
print('Best Score :', grid_result.best_score_)
print('Best Parameters :', grid_result.best_params_)
```

```python
# Stores the optimum model in best_pipe
best_pipe = grid_result.best_estimator_
print(best_pipe)
```

```python
from pandas import DataFrame

result_df = DataFrame.from_dict(grid_result.cv_results_, orient='columns')
print(result_df.columns)
```

## training the model with best params

```python
def create_model(optimizer='adam', kernel='uniform'):
    model = Sequential()
    model.add(Dense(units=60, input_dim=60, kernel_initializer=kernel, activation='relu', name='Hidden-Layer-1'))
    model.add(Dense(units=1, kernel_initializer=kernel, activation='sigmoid', name='Output-Layer-1'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
# check point

# filepath="weights.{epoch:02d}-{accuracy:.2f}.hdf5"
filepath = "sonar-best-modal.hdf5"

# Keep only a single checkpoint, the best over test accuracy.
checkpoint = ModelCheckpoint(filepath = filepath,
                             monitor='accuracy',
                             verbose=1,
                             save_best_only=True,
                             save_weights_only=False,
                             save_freq="epoch",
                             mode='max')
callbacks_list = [checkpoint]
```

```python
final_kears_estimator= KerasClassifier(build_fn=create_model, 
                                       epochs=30, 
                                       batch_size=5, 
                                       validation_split=0.33
                                       verbose=1,
                                       callbacks=callbacks_list
                                      )

pipelines = Pipeline([('ss', StandardScaler()),                     
                     ("kc", final_kears_estimator)])  
```

```python

start_time = time.time()

final_pipeline = pipelines.fit(X_train,y_train)
    
print(final_pipeline)


training_time = time.time() - start_time
print('\n Total Training Time taken :', convert(training_time))
```

```python
print('Training set score: ' + str(final_pipeline.score(X_train,y_train)))
print('Test set score: ' + str(final_pipeline.score(X_test,y_test)))
```
